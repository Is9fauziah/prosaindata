```python
import csv
import requests
from bs4 import BeautifulSoup

# Membuat file CSV dan menulis header
with open("hasil.csv", "w", newline="", encoding="utf-8") as csvfile:
    writer = csv.writer(csvfile)
    writer.writerow(["Judul", "Link Abstrak", "Abstrak"])

    for i in range(1, 172):
        url = f"https://pta.trunojoyo.ac.id/c_search/byprod/10/{i}"
        response = requests.get(url)
        soup = BeautifulSoup(response.content, "html.parser")
        results = soup.find(id="content")
        karya_ilmiah_elements = results.find_all(attrs={"data-id": True})

        for karya_ilmiah_element in karya_ilmiah_elements:
            judul_element = karya_ilmiah_element.find("a", class_="title")
            abstrak_url = karya_ilmiah_element.find("a", class_="gray button").get("href")
            
            # request ke halaman abstrak
            abstrak_response = requests.get(abstrak_url)
            abstrak_soup = BeautifulSoup(abstrak_response.content, "html.parser")
            abstrak = abstrak_soup.find("p")
            
            # Menulis baris data ke file CSV
            writer.writerow([judul_element.text.strip(), abstrak_url, abstrak.text.strip()])
            print(judul_element.text.strip())
            #print(abstrak_url)
            #print(abstrak.text.strip)
            print()

print("Data berhasil disimpan ke file hasil.csv")
```
