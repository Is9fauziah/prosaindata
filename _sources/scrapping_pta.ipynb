```python
import csv
import requests
import json
from bs4 import BeautifulSoup


data = []
for i in range(1, 172):
    url = f"https://pta.trunojoyo.ac.id/c_search/byprod/10/{i}"
    response = requests.get(url)
    soup = BeautifulSoup(response.content, "html.parser")
    results = soup.find(id="content")
    karya_ilmiah_elements = results.find_all(attrs={"data-id": True})

    for karya_ilmiah_element in karya_ilmiah_elements:
        judul_element = karya_ilmiah_element.find("a", class_="title")
        penulis_element = karya_ilmiah_element.find_all("div")[0].find("span").text
        dosen1_element = karya_ilmiah_element.find_all("div")[2].find("span").text
        dosen2_element = karya_ilmiah_element.find_all("div")[3].find("span").text
        abstrak_url = karya_ilmiah_element.find("a", class_="gray button").get("href")

        # request ke halaman abstrak
        abstrak_response = requests.get(abstrak_url)
        abstrak_soup = BeautifulSoup(abstrak_response.content, "html.parser")
        abstrak = abstrak_soup.find("p").text.strip()

        # store the data as dictionary
        item = {
            "judul": judul_element.text.strip(),
            "penulis": penulis_element,
            "dosen1": dosen1_element,
            "dosen2": dosen2_element,
            "abstrak_url": abstrak_url,
            "abstrak": abstrak
        }
        data.append(item)

# convert the data to JSON format and write to file
with open("hasil.json", "w") as file:
    json.dump(data, file, indent=4)

print("Data berhasil disimpan ke file hasil.json")
```
